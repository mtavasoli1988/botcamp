```python
A = np.array(((2, 1), (1, 3)))
B = np.array(((-2, 4.2), (4.2, 6)))
C = np.dot(A, B) # matrix-matrix product
print(C)
```

```python
w, v = np.linalg.eig(A)
print(w,v)
```

```python
print(w)
```

```python
print(v)
```

```python
b = np.array((1, 2))
x = np.linalg.solve(C, b)
print(x)
```

```python
print(np.dot(C, x))
```
